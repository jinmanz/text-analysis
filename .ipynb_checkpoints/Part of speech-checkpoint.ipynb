“Parts of speech are linguistic categories that identify the syntactic function of words, such as determiners (or articles), nouns, verbs, adjectives, prepositions, etc. Imagine, for instance, that we could determine that Text A has more adjectives per total number of words than Text B – might that allow us to speculate that Text A is more descriptive than Text B?”

### Taging parts of speech

```python
import nltk
# getting txt

with open("Plaintexts/Adrift in New York.txt","r") as f:
    Atext = f.read()
Atokens = nltk.word_tokenize(Atext)
```

"we'll initially use the simplest syntax possible to tag the parts of speech in our tokens list"

```python
Atagged = nltk.pos_tag(Atokens)
```

```python
# peek at the first 20 token-tag pairs
Atagged[:20]
```

- DT: determiner (the, a…)
- IN: preposition (to, from, …)
- JJ: adjective (big, smart…)
- NN: noun (car, school…)
- NNP: proper noun (Montreal, Jill…)
- PRP: personal pronoun (she, I…)
- RB: adverb (freely, calmly…)
- VBD, VBG, VBN, VBZ: verbs (dancing, said, been, is…)
- WP: wh-pronoun (who, what…)


### Frequencies and distribution of a specific part of speech

```python
Aadj = [token for token, pos in Atagged if "JJ" == pos]

# nltk.FreqDist(Aadj) return a FreqDist object
AadjFreqs = nltk.FreqDist(Aadj)

# plot the top 20 adj
AadjFreqs.plot(20)
```

```python
AgoldBugProperNouns  = [token for token,pos in Atagged if "NNP" in pos]
nltk.FreqDist(AgoldBugProperNouns).plot(20)
```

### Overall Frequencies of Part of Speech

```python
%matplotlib inline
Atags = [pos for word, pos in Atagged]
AtagsFreqs = nltk.FreqDist(Atags)
AtagsFreqs.tabulate(25)
AtagsFreqs.plot(20)
```
